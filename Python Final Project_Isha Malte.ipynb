18.Population: Integrer- Population of the country  
19.thinness 1-19 years: Integrer- Prevalence of thinness among children and adolescents for Age 10 to 19 (% )  
20.thinness 5-9 years: Integrer- Prevalence of thinness among children for Age 5 to 9(%)

### Finding and handling null values

```python
#to know the percentage of null values in each column. This will help in deciding how to handle the null values
life.isnull().sum()*100/len(life)
```

Hepatitis B, GDP and Population have maximum and a high no. of null values. As these could be important columns in analysis, I will not delete the rows having their null values.  
Filling with a generic mean or any arbitary value will not be correct as this data differes largely for country to country. Hence, I have used mean of that particular field for that particular country.

```python
#Fill each null value with average of that column specific to the country. 
# If some category has only NaN values, so use mean of all values of column for filling NaN

life['Hepatitis B'] = life.groupby('Country')['Hepatitis B'].apply(lambda x: x.fillna(x.mean()))
life['Hepatitis B'] = life['Hepatitis B'].fillna(life['Hepatitis B'].mean())

life['Life expectancy'] = life.groupby('Country')['Life expectancy'].apply(lambda x: x.fillna(x.mean()))
life['Life expectancy'] = life['Life expectancy'].fillna(life['Life expectancy'].mean())

life['Adult Mortality'] = life.groupby('Country')['Adult Mortality'].apply(lambda x: x.fillna(x.mean()))
life['Adult Mortality'] = life['Adult Mortality'].fillna(life['Adult Mortality'].mean())

life['BMI'] = life.groupby('Country')['BMI'].apply(lambda x: x.fillna(x.mean()))
life['BMI'] = life['BMI'].fillna(life['BMI'].mean())

life['Total expenditure'] = life.groupby('Country')['Total expenditure'].apply(lambda x: x.fillna(x.mean()))
life['Total expenditure'] = life['Total expenditure'].fillna(life['Total expenditure'].mean())

life.Alcohol = life.groupby('Country')['Alcohol'].apply(lambda x: x.fillna(x.mean()))
life.Alcohol = life.Alcohol.fillna(life.Alcohol.mean())

life['Polio'] = life.groupby('Country')['Polio'].apply(lambda x: x.fillna(x.mean()))
life['Polio'] = life['Polio'].fillna(life['Polio'].mean())

life['Diphtheria'] = life.groupby('Country')['Diphtheria'].apply(lambda x: x.fillna(x.mean()))
life['Diphtheria'] = life['Diphtheria'].fillna(life['Diphtheria'].mean())

life['GDP'] = life.groupby('Country')['GDP'].apply(lambda x: x.fillna(x.mean()))
life['GDP'] = life['GDP'].fillna(life['GDP'].mean())

life['Population'] = life.groupby('Country')['Population'].apply(lambda x: x.fillna(x.mean()))
life['Population'] = life['Population'].fillna(life['Population'].mean())

life['thinness  1-19 years'] = life.groupby('Country')['thinness  1-19 years'].apply(lambda x: x.fillna(x.mean()))
life['thinness  1-19 years'] = life['thinness  1-19 years'].fillna(life['thinness  1-19 years'].mean())

life['thinness 5-9 years'] = life.groupby('Country')['thinness 5-9 years'].apply(lambda x: x.fillna(x.mean()))
life['thinness 5-9 years'] = life['thinness 5-9 years'].fillna(life['thinness 5-9 years'].mean())

life['Income composition of resources'] = life.groupby('Country')['Income composition of resources'].apply(lambda x: x.fillna(x.mean()))
life['Income composition of resources'] = life['Income composition of resources'].fillna(life['Income composition of resources'].mean())

life['Schooling'] = life.groupby('Country')['Schooling'].apply(lambda x: x.fillna(x.mean()))
life['Schooling'] = life['Schooling'].fillna(life['Schooling'].mean())
```

### Min-Max Scaling (Normalization) and Outlier Handling
As many features are not normal and I would like to model it with linear regression, we have to perform normalisation. I chose this because there are many outliers present and minmax will reduce the variance and effect of the outllier

```python
#Create the min-max object
from sklearn import preprocessing
minmax_scaler = preprocessing.MinMaxScaler()

# apply the transformation to the training data
X_train_minmax = minmax_scaler.fit_transform(xTrain) 

# apply the transformation to the testing data
X_test_minmax = minmax_scaler.transform(xTest) # but we only transform our testing data with already fit scaler

# convert resulting array back to dataframe
Xtest_minmax = pd.DataFrame(X_test_minmax,
                     columns=xTrain.columns)

Xtest_minmax.head()
```

```python
Xtest_minmax.describe()
```

We see that the min and max values are sometimes slightly larger than 1 and smaller than 0. This is because as per the formula of min-max ((X-Xmin)/(Xmax-Xmin)) this is possible for our due to outliers. But this outlier effect has been minimised by the minmax normalisation

## 4. Model Implementation

### 1. Multiple Regression Analysis
I implemendted this model because I wanted to check the performance on a simple model first and the below assumptions for this model are satisfied:  
Linear relationship  
Multivariate normality  
No or little multicollinearity  
No auto-correlation  

```python
#Multiple linear regression
from sklearn import linear_model

# instantiate the model object
lm = linear_model.LinearRegression()

# fit the model to the dataset
lm_model = lm.fit(xTrain, yTrain)

# make predictions
linear_preds = lm.predict(xTest)
print(linear_preds)
```

```python
df = pd.DataFrame(yTest[:])
df["preds"] = linear_preds
df.head()
```

```python
coeff_df = pd.DataFrame(lm.coef_.T, xTrain.columns, columns=['Coefficient'])  
coeff_df
```

Status developed and developing are contributing in positive and negative way to life expectancy respectively.

### 2. Lasso Regression
To minimize the weight of coefficients by checking on multiple combinations of weights and iteration and generting the best possible R square. I performed a grid search with cross-validation on the grid regression with alpha varying between 0 and 1 and having 3 max iterations of 10, 100 and 1000.

```python
#Lasso regression
from sklearn.metrics import make_scorer
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

grid_cvl = GridSearchCV(Lasso(),
              param_grid={'alpha': [0.001,0.005,0.01,0.05,0.1,0.5,1], 'max_iter' : [10, 100, 1000]},
              scoring='r2', cv=5, refit=True)


grid_cvl.fit(xTrain, yTrain)
print("Best Parameters: " + str(grid_cvl.best_params_))

result1 = grid_cvl.cv_results_
```

```python
coeff_df_lasso = pd.DataFrame(lasso_mod.coef_.T, xTrain.columns, columns=['Coefficient'])  
coeff_df_lasso
```

5.Income composition of resources is positively correlated (0.69). By increaing the income of the citizens, their standard of living and awareness also increases, thus increasing no. of years lived.  
  
There is a linear relationship between below features and life expectancy which means that increasing one will also increase life expectancy and vice versa:  
1.adult mortality  
2.income composition of resources  
3.schooling  
Global authorities and every country's government should take help of these 3 factors as they have the simplest correlation to increase life expectancy.  
  
Population is no way related to life expdectancy or any other feature. Thus, a myth that densely populated countries have lower life expectancy is incorrect.
